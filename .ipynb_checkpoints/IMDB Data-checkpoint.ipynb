# IMDB Data

```python
import pandas as pd
import numpy as np
import json 
import jsonlines
```

```python
response=[]
with jsonlines.open('IMDB/title_essential_v1_sample.jsonl') as f:

    for line in f.iter():

        response.append(line)
```

```python
imdb = pd.DataFrame.from_dict(response)
```
