```python
import twint
import nest_asyncio
import pandas as pd

nest_asyncio.apply()
```

```python
data = pd.read_csv('twitter_tags.csv', lineterminator='\n')
```

```python
twitter_tags = pd.DataFrame(data)
twitter_tags.drop(['Unnamed: 0'], axis = 1, inplace = True)
twitter_tags
```

```python
twitter_tags = list(twitter_tags['0'])
```

```python
twitter_tags = [tweet.split('^^') for tweet in twitter_tags]
```

```python
twitter_tags
```

```python
tweet_counter = []
for tweet in twitter_tags:
    c = twint.Config()
    c.Search = tweet[0]
    c.Since = tweet[1]
    c.Until = tweet[2]
    c.Pandas = True
    c.Hide_Output = True
    
    twint.run.Search(c)
    df = twint.storage.panda.Tweets_df
    tweet_counter.append([len(df)])
```

```python
tweet_counter = []
c = twint.Config()
c.Search = "Avengers"
c.Since = "2019-01-01"
c.Until = "2019-01-02"
c.Pandas = True
c.Hide_Output = True

twint.run.Search(c)
df = twint.storage.panda.Tweets_df
tweet_counter.append([len(df)])
```
